```python
import numpy as np
import matplotlib.pyplot as plt

# 假设 folder_path 是包含文件的目录
folder_path = './'

# 加载文件
predictions = np.load(f'{folder_path}pred.npy')
true_values = np.load(f'{folder_path}true.npy')

# 指定样本索引和特征索引
sample_index = 30  # 更改这个索引来查看不同的样本
feature_index = sample_index  # 您关注的特征的索引


# 提取特定样本的特定特征的预测值和真实值
pred_feature = predictions[sample_index, :, feature_index]
true_feature = true_values[sample_index, :, feature_index]

# 选择绘制的数据量
num_points_to_plot = 20  # 例如，绘制前20个时间点

# 绘制折线图
plt.figure(figsize=(11, 6))  # 设置图形大小
plt.plot(true_feature[:num_points_to_plot], label='True Values', marker='o')
plt.plot(pred_feature[:num_points_to_plot], label='Predictions', marker='x')
plt.title(f'Feature {feature_index} - Sample {sample_index}')
plt.xlabel('Time Steps')
plt.ylabel('Value of Feature')
plt.legend()
plt.show()
```
